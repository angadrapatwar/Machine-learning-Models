**Step 1: Importing the libraries**

```python
import numpy as np
import pandas as pd
```

**Step 2: Importing dataset**

```python
df=pd.read_csv("Data.csv")
```

**Step 3: Handling the missing data**

```python
# handling the missing values in the age column
mean=round(df["Age"].mean())
mean
```

```python
df["Age"].fillna(value=mean, inplace=True)
```

```python
mean1= round(df["Salary"].mean())
mean1
```

```python
df["Salary"].fillna(value=mean1, inplace=True)
```

```python
df
```

**Step 4: Encoding categorical data**

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_x= LabelEncoder()  
df['Purchased']= labelencoder_x.fit_transform(df["Purchased"])
```

**Step 5: Creating a dummy variable**

```python
df1=pd.get_dummies(df["Country"])
```

```python
df1
```

```python
df = pd.concat([df1, df], axis=1)
```

```python
df
```

```python
df.drop("Country", axis=1, inplace=True)
```

```python
df
```

```python
#as the first three columns are already in binar form we will convert only the age and salary column
from sklearn.preprocessing import StandardScaler
st_x= StandardScaler()    
df['Age']= st_x.fit_transform(df[['Age']])
```

```python
df['Salary']= st_x.fit_transform(df[['Salary']])
```
