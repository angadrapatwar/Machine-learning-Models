### **Splitting the dataset**

```python
#first splitting the data set into independent and dependent data
x=df.iloc[:,:9].values         #independent 
y=df.iloc[:,9].values          #dependent
```

```python
#now splittng this dependent and independent data into training and test set

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.20, random_state=None)
```

```python
#scaling only the independent variables 

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#using distance metrics as square euclidean
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=10, p=2, metric='euclidean')

knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
#using distance metrics as manhattan distance
knn = KNeighborsClassifier(n_neighbors=10, p=1, metric='manhattan')

knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```
